```python
import tensorflow as tf
tf.config.run_functions_eagerly(True)
a=tf.constant('hello world')
print(a.numpy())
```

```python
#scalar constants
a=tf.constant(42)
b=tf.constant(3.12)
c=tf.constant('hello,tensorflow!')
print(a)
print(b)
print(c)
```

```python
#vector constants
a=tf.constant([1,2,3])
b=tf.constant([1.0,2.0,3.0])
print(a)
print(b)
print(a.numpy())
print(b.numpy())
```

```python
#matrix constants
a=tf.constant([[1,2],[3,4]])
print(a)
print(a.numpy())
```

```python
#higher dimenssion matrix constants
a=tf.constant([[[1,2],[3,4]],[[1,2],[3,4]]])
#print(a)
print(a.numpy())
```

```python
a=tf.constant(True)
print(a)
print(a.numpy())
```

```python
a=tf.zeros(shape=(3,3))
print(a)
print(a.numpy())
```

```python
a=tf.ones(shape=(3,3))
print(a)
print(a.numpy())
```

```python
a=tf.eye(3)
print(a)
print(a.numpy())
```

```python
a=tf.constant([[1,2],[3,4]])
b=tf.constant([[2,3],[4,5]])
c=tf.constant(4)
print((a*b).numpy())
print((a*c).numpy())
print((tf.matmul(a,b)).numpy())
print((a@b).numpy())
```

```python
a=tf.constant(1.25)
print(a.numpy())
b=tf.cast(a,dtype=tf.int32)
print(b.numpy())
```

```python
x=tf.constant([[1.,2.,3.],[4.,5.,6.]])
f=tf.concat([x,x,x],axis=1)
print(f.numpy())
f1=tf.concat([x,x,x],axis=0)
print(f1.numpy())
```

```python
#x=tf.constant([[11.0,12.0,13.0]])
logits=tf.constant([20.0,30.0,40.0,60.0,70.0])
softmax_probs=tf.nn.softmax(logits,axis=-1)
print((softmax_probs).numpy())
```
