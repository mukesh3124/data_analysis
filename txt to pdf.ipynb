```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/WhatsApp Chat with Subbalachimi.txt')
df
df=df.drop(df.columns[2],axis=1)
df=df.drop(df.columns[2],axis=1)
df.columns=['date','msg']

import matplotlib.pyplot as plt
```

```python
import matplotlib.backends.backend_pdf as pdf_backend
```

```python
pdf_filename = 'output.pdf'
```

```python
with pdf_backend.PdfPages(pdf_filename) as pdf:
    fig, ax = plt.subplots(figsize=(8, 4))  # Adjust the figure size if needed
    ax.axis('off')  # Turn off axis labels
    table_data = pd.plotting.table(ax, df, loc='center', colWidths=[1] * len(df.columns))
    table_data.auto_set_font_size(False)
    table_data.set_fontsize(10)
    table_data.scale(1.2, 1.2)  # Adjust the scaling if needed
    pdf.savefig(fig, bbox_inches='tight')
    plt.close()
```
