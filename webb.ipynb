```python
b='https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1'
page=requests.get(b)
```

```python
h_code1=BeautifulSoup(page.text,'html')
print(h_code1)
```

```python
t1=h_code1.find('table')
print(t1)
```

```python
t2=t1.find_all('th')
print(t2)

```

```python
head=[]
for i in t2:
 head.append(i.text)
print(head)
```

```python
#list name=[o/p   for loop]
head1=[i.text for i in t2]
print(head1)
```

```python
#creating a data frame
import pandas as pd
df=pd.DataFrame(columns=head1)
print(df)
```

```python
rows=t1.find_all('tr')
#print(rows)
ind=0
for i in rows[1:6]:
  erd=i.find_all('td')
  rd=[j.text for j in erd]
  print(rd)
  df.loc[ind]=rd
  ind=ind+1
```
